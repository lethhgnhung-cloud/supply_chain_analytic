```python
df_processed = df.copy()
print("DataFrame đã được sao chép.")
```

### 1. Xử lý Outliers với RobustScaler


```python
from sklearn.preprocessing import RobustScaler

# Các cột có outliers
r_scaler_cols = ['Price', 'Revenue generated', 'Stock levels', 'Shipping costs']

# Khởi tạo RobustScaler
robust_scaler = RobustScaler()

# Áp dụng RobustScaler
df_processed[r_scaler_cols] = robust_scaler.fit_transform(df_processed[r_scaler_cols])

print("Đã áp dụng RobustScaler cho các cột:\n", r_scaler_cols)
```

### 2. Chuẩn hóa với StandardScaler (Z-score normalization)


```python
from sklearn.preprocessing import StandardScaler

# Lấy tất cả các cột số ban đầu
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Xác định các cột đã được RobustScaler xử lý
scaled_by_robust_cols = set(r_scaler_cols)

# Các cột còn lại để áp dụng StandardScaler
s_scaler_cols = [col for col in numerical_cols if col not in scaled_by_robust_cols]

# Loại bỏ 'SKU' nếu nó xuất hiện trong numerical_cols (mặc dù nó là object)
# Tuy nhiên, SKU là một định danh duy nhất và không nên được đưa vào scaling.
# df.info() cho thấy SKU là 'object', nên nó sẽ không có trong numerical_cols. An toàn để bỏ qua.

# Khởi tạo StandardScaler
standard_scaler = StandardScaler()

# Áp dụng StandardScaler
df_processed[s_scaler_cols] = standard_scaler.fit_transform(df_processed[s_scaler_cols])

print("Đã áp dụng StandardScaler cho các cột:\n", s_scaler_cols)
```

### 3. Mã hóa biến phân loại (Categorical Variables) với One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

# Các cột phân loại cần mã hóa (loại bỏ 'SKU')
categorical_cols = df.select_dtypes(include='object').columns.tolist()
if 'SKU' in categorical_cols:
    categorical_cols.remove('SKU')

# Khởi tạo OneHotEncoder
# handle_unknown='ignore' để xử lý các giá trị không thấy trong quá trình fit
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Áp dụng OneHotEncoder
encoded_features = one_hot_encoder.fit_transform(df_processed[categorical_cols])

# Tạo DataFrame từ các đặc trưng đã mã hóa
encoded_df = pd.DataFrame(encoded_features, columns=one_hot_encoder.get_feature_names_out(categorical_cols), index=df_processed.index)

# Xóa các cột phân loại gốc khỏi df_processed
df_processed = df_processed.drop(columns=categorical_cols)

# Nối các đặc trưng đã mã hóa vào df_processed
df_processed = pd.concat([df_processed, encoded_df], axis=1)

print("Đã áp dụng One-Hot Encoding cho các cột:\n", categorical_cols)
print("Kích thước DataFrame sau mã hóa:", df_processed.shape)
display(df_processed.head())
```

### Phase 3: SKU scoring dashboard

### 1. Tính toán điểm số SKU

```python
# Tùy chỉnh trọng số cho các yếu tố khác nhau
# Các yếu tố được chọn từ df_processed (đã được chuẩn hóa)
# Ví dụ: Revenue generated (doanh thu), Defect rates (tỷ lệ lỗi), Stock levels (mức tồn kho), Lead times (thời gian chờ)

# Tạo một bản sao của df_processed để thêm cột điểm
df_scoring = df_processed.copy()

# Các trọng số (có thể điều chỉnh):
# Trọng số dương: muốn tối đa hóa (VD: doanh thu, số lượng bán)
# Trọng số âm: muốn tối thiểu hóa (VD: tỷ lệ lỗi, chi phí)
w_revenue = 0.4
w_defect = -0.3
w_stock = 0.1
w_price = 0.1
w_lead_time = -0.1 # Sử dụng Manufacturing lead time do có nhiều giá trị hơn
w_prod_vol = 0.1 # Khối lượng sản xuất

# Tính điểm
# Sử dụng các cột đã được chuẩn hóa
# SKU là một định danh, không nên dùng để tính điểm

df_scoring['Score'] = (
    df_scoring['Revenue generated'] * w_revenue +
    df_scoring['Defect rates'] * w_defect +
    df_scoring['Stock levels'] * w_stock +
    df_scoring['Price'] * w_price +
    df_scoring['Manufacturing lead time'] * w_lead_time +
    df_scoring['Production volumes'] * w_prod_vol
)

print("Đã tính toán điểm số cho các SKU.")
display(df_scoring[['SKU', 'Score']].head())
```

### 2. Chuẩn hóa điểm số (ví dụ: về thang 0-100)

```python
from sklearn.preprocessing import MinMaxScaler

# Khởi tạo MinMaxScaler
min_max_scaler = MinMaxScaler(feature_range=(0, 100))

# Áp dụng MinMaxScaler cho cột 'Score'
df_scoring['Normalized_Score'] = min_max_scaler.fit_transform(df_scoring[['Score']])

print("Đã chuẩn hóa điểm số về thang 0-100.")
display(df_scoring[['SKU', 'Score', 'Normalized_Score']].head())
```

### 3. Xếp hạng và Trực quan hóa các SKU hàng đầu

```python
# Sắp xếp theo điểm số đã chuẩn hóa giảm dần
top_skus = df_scoring.sort_values(by='Normalized_Score', ascending=False).head(10)

print("Top 10 SKU có điểm cao nhất:")
display(top_skus[['SKU', 'Normalized_Score', 'Product type_cosmetics', 'Product type_haircare', 'Product type_skincare']])

# Trực quan hóa Top SKU
plt.figure(figsize=(12, 7))
sns.barplot(x='SKU', y='Normalized_Score', data=top_skus, palette='viridis', hue='SKU', legend=False)
plt.title('Top 10 SKU theo Normalized Score')
plt.xlabel('SKU')
plt.ylabel('Normalized Score (0-100)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
